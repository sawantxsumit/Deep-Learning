```python
from tensorflow.keras.layers import Embedding ,Input
from tensorflow.keras.models import Sequential
from tf_keras.utils import pad_sequences
from tf_keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
# vocab size
voc_size=5000
```

```python
messages=x.copy()
```

```python
messages
```

```python
messages['title'][0]
```

```python
messages.reset_index(inplace=True)
```

```python
messages
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
nltk.download('wordnet')
```

```python
# Dataset prepocessing
```

```python
from nltk.stem import WordNetLemmatizer
lmt=WordNetLemmatizer()
```

```python
corpus=[]
```

```python
for i in range(0, len(messages)):
  # print(i)
  review=re.sub('[^a-zA-Z]', ' ', messages['title'][i])
  review=review.lower()
  review=review.split()

  review=[lmt.lemmatize(word) for word in review if not word in stopwords.words('english')]
  review=" ".join(review)
  corpus.append(review)
```

```python
corpus
```

```python
len(corpus)
```

```python
# One hot encoding
```

```python
onehot_repr=[one_hot(words, voc_size) for words in corpus]
onehot_repr
```

```python
corpus[1]
```

```python
onehot_repr[1]
```

```python
# Padding
```

```python
len(max(corpus))
```

```python
sentence_length= 60
embedded_docs= pad_sequences(onehot_repr, padding='pre', maxlen=sentence_length)
print(embedded_docs)
```

# Creating LSTM model

```python
embedding_vector_features=40
model=Sequential()
model.add(Input(shape=(sentence_length,)))
model.add(Embedding(voc_size, embedding_vector_features))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np
x_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
x_final.shape , y_final.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain , ytest = train_test_split(x_final,y_final , test_size=0.25 , random_state=33)
```

```python
# Model training
```

```python
print(xtrain.dtype, ytrain.dtype)
print(type(xtrain), type(ytrain))
```

# Adding dropouts

```python
from tensorflow.keras.layers import Dropout
embedding_vector_features=40
model=Sequential()
model.add(Input(shape=(sentence_length,)))
model.add(Embedding(voc_size, embedding_vector_features))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(xtrain[:5000] , ytrain[:5000] , validation_data=(xtest , ytest), epochs=10 , batch_size=64)
```

```python
import pickle
pickle.dump(model, open('lstm.pkl', 'wb'))
```

```python
y_pred=model.predict(xtest)
```

```python
y_pred=np.where(y_pred > 0.5, 1, 0)
```

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix , roc_auc_score
```

```python
print("Accuracy score :" , accuracy_score(ytest , y_pred))
print("Confusion matrix :" , confusion_matrix(ytest , y_pred))
print("Classification report :" , classification_report(ytest , y_pred))
```
