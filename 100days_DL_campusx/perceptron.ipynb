```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("placement.csv")
```

```python
df.head()
```

```python
sns.scatterplot(x=df['cgpa'], y=df['resume_score'] , hue=df['placed'])
```

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p= Perceptron()
```

```python
p.fit(x,y)
```

```python
p.coef_
# w1= -44.9 and w2=51.72
```

```python
p.intercept_
# here b= -25
```
