```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# !pip install scikit-learn
```

```python
housing= fetch_california_housing()
```

```python
# Train test split
xtrain_full , xtest , ytrain_full , ytest = train_test_split(housing.data , housing.target)
```

```python
# Training validation split
xtrain , xvalid , ytrain , yvalid= train_test_split(xtrain_full , ytrain_full)
```

```python
scaler = StandardScaler()
xtrain_scaled= scaler.fit_transform(xtrain)
xvalid_scaled= scaler.transform(xvalid)
xtest_scaled= scaler.transform(xtest)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model= keras.models.Sequential()
model.add(keras.layers.Dense(30 , activation='relu' ,input_shape=xtrain.shape[1:]))
model.add(keras.layers.Dense(1))
```

```python
model.compile(loss='mean_squared_error' , optimizer='sgd')
```

```python
model.summary()
```

```python
history= model.fit(xtrain , ytrain ,validation_data=(xvalid , yvalid))
```

```python
mse_test= model.evaluate(xtest , ytest)
xnew= xtest[:3]
```

```python
ypred= model.predict(xnew)
```
